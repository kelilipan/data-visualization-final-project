# Dataprep

```python
df_confirmed = pd.read_csv('data/time_series_covid19_confirmed_global.csv')
df_death = pd.read_csv('data/time_series_covid19_deaths_global.csv')
df_recovered = pd.read_csv('data/time_series_covid19_recovered_global.csv')
```

```python
regions_death = df_death['Country/Region']
regions_confirmed = df_confirmed['Country/Region']
regions_recovered = df_recovered['Country/Region']
```

```python
df_death = df_death.drop(['Province/State','Lat','Long','Country/Region'],axis=1).T
df_death.columns = regions_death
df_death.to_csv('data/death.csv')
```

```python
df_confirmed = df_confirmed.drop(['Province/State','Lat','Long','Country/Region'],axis=1).T
df_confirmed.columns = regions_confirmed
df_confirmed.to_csv('data/confirmed.csv')
```

```python
df_recovered= df_recovered.drop(['Province/State','Lat','Long','Country/Region'],axis=1).T
df_recovered.columns = regions_recovered
df_recovered.to_csv('data/recovered.csv')
```
